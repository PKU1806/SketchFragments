```python
# -*- coding: UTF-8 -*-
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.rc('font', family='Times New Roman')
```

```python
# FE - ARE

x_axix = np.array([16*i for i in range(2,12)])
# x_axix[0] = 50

res = [[] for i in range(3)]

#开始画图
plt.figure(figsize=(6, 4.5))
# plt.title('Frequency Estimation', fontsize = 13)
# plt.xlim( 2**(2.5),  2**13.5 )
plt.xscale('linear')
plt.yscale('linear')
plt.tick_params(labelsize = 20)

with open('../log/log_fe_mininet10k') as f:
    cnt = 0
    for line in f:
        data = [float(x) for x in line.split()]
        res[cnt % 3].append(data[1]) # are
        cnt += 1
        #if cnt > 3000:
        #    break

x_axix = x_axix
plt.plot(x_axix, res[0], label='Half CU(ours)', linestyle='-', marker='s', markerfacecolor='none', alpha = 0.6)
plt.plot(x_axix, res[1], label='CM Sketch', linestyle='-', marker='^', markerfacecolor='none', alpha = 0.6)
plt.plot(x_axix, res[2], label='CU Sketch', linestyle='-', marker='v', markerfacecolor='none', alpha = 0.6)
# plt.plot(x_axix, res[7], label='Heavy-Guardian', linestyle='-', marker='x', alpha = 0.8)
plt.legend(loc = 'upper right', fontsize = 19) # 显示图例

plt.grid(True, linestyle='--', axis='y')

print(plt.ylim())
# plt.ylim(top=2500)
plt.xlabel('Number of Bucket', fontsize = 26)
plt.ylabel('ARE', fontsize = 26)
plt.tight_layout()
plt.savefig('fe_are_mininet10k.png', ppi=400)
plt.show()
#python 一个折线图绘制多个曲线
```
